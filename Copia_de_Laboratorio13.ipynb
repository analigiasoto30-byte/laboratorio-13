Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: "The system design exceeded expectations and made the purchasing decision easy".
Frase 2: "The technical tests failed repeatedly, making this purchase a poor choice".
Frase 3: "The project evaluation meets the basic requirements for this purchase"."
Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

# Frases a analizar
frases = [
    "The system design exceeded expectations and made the purchasing decision easy",
    "The technical tests failed repeatedly, making this purchase a poor choice",
    "The project evaluation meets the basic requirements for this purchase"
]

# Función para clasificar el sentimiento
def clasificar_sentimiento(polaridad):
    if polaridad > 0:
        return "Positivo"
    elif polaridad < 0:
        return "Negativo"
    else:
        return "Neutral"

# Análisis de cada frase
for i, frase in enumerate(frases, 1):
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity
    sentimiento = clasificar_sentimiento(polaridad)

    print(f"Frase {i}: {frase}")
    print(f"Polaridad: {polaridad}")
    print(f"Sentimiento: {sentimiento}")
    print("-" * 50)
```

Frase 1: "The system design exceeded expectations and made the purchasing decision eas"
Traducción: "El diseño del sistema superó las expectativas y facilitó la decisión de compra."
Frase 2: "The technical tests failed repeatedly, making this purchase a poor choice."
Traducción: "Las pruebas técnicas fallaron repetidamente, haciendo que esta compra sea una mala elección."
Frase 3: "The project evaluation meets the basic requirements for this purchase."
Traducción: "La evaluación del proyecto cumple con los requisitos básicos para esta compra."

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
import nltk
nltk.download('punkt_tab')

from textblob.classifiers import NaiveBayesClassifier

# Datos de entrenamiento (frase, etiqueta)
train_data = [
    ("This app is very intuitive!", "pos"),
    ("I didn't enjoy the customer service.", "neg"),
    ("Quite average experience.", "neu"),
    ("Fantastic presentation!", "pos"),
    ("The food was awful.", "neg")
]

# Crear el clasificador
cl = NaiveBayesClassifier(train_data)

# Evaluación del modelo con los mismos datos (ejemplo básico)
accuracy = cl.accuracy(train_data)

print(f"Precisión del modelo: {accuracy:.2f}")
print("-" * 50)

# Probar el modelo con las frases
test_sentences = [
    "This app is very intuitive!",
    "I didn't enjoy the customer service.",
    "Quite average experience.",
    "Fantastic presentation!",
    "The food was awful."
]

for sentence in test_sentences:
    prediction = cl.classify(sentence)
    prob_dist = cl.prob_classify(sentence)

    print(f"Frase: {sentence}")
    print(f"Clasificación: {prediction}")
    print(f"Probabilidad positiva: {prob_dist.prob('pos'):.2f}")
    print(f"Probabilidad negativa: {prob_dist.prob('neg'):.2f}")
    print("-" * 50)
```
